## Multiclass Datasets - Overview:
It is expected that mutliclass datasets are not so well measured by accuracy. Instead, we can use something like roc_auc_ovr. Different of roc_auc, the roc_auc_ovr is a good measure to multiclass dataset because it measures each value of target against the all other targets.

```python
for n in [3, 5, 7, 9, 11, 13, 15]:
    knn_metrics_multiclass_n = knn_metrics.query("tasks_description == 'multiclass_tasks' & neighbours == @n")
    
    acc_weight_means = knn_metrics_multiclass_n.groupby('weights')['accuracy'].mean()
    print(f"\nMean accuracy for {n} neighbours:", uniform_weight_means)

    roc_weight_means = knn_metrics_multiclass_n.groupby('weights')['roc_auc_ovr'].mean()
    print(f"\nMean roc_auc for {n} neighbours:", roc_weight_means)
```

```python
def is_numerical(column):
    try:
        if column.dtype in (np.float64, np.int64, np.uint8):
            return True
        return False
    except ValueError:
        return False
```

```python
from pandas.api.types import is_categorical_dtype

def is_categorical(column):
    try:
        if column.dtype in (object, bool) or is_categorical_dtype(column):
            return True
        return False
    except ValueError:
        return False
```

### Conclusions:

The analyzed datasets, which are considered imbalanced (with the minority class normalized <= 0.3), seemingly posed challenges for KNN to yield good roc_auc metrics due to the presence of numerous noisy columns. It was observed that high correlation between the target class and individual columns is not common, but there are sometimes strong interactions among the columns themselves, which significantly influences the dataset. Consequently, in dataset 3904, for instance, most columns are correlated with each other.

The distributions of values appear to be skewed in some cases.

Perhaps changes in how noisy columns are computed could yield better results.

```python
numeric_cols = [column for column in dataset_31.columns if is_numerical(dataset_31[column])]
numeric_cols_with_class = numeric_cols + ['class']
dataset_31['class'] = dataset_31['class'].replace({'good': 1, 'bad': 2})
correlation_matrix = dataset_31[numeric_cols_with_class].corr(method='pearson')
plt.figure(figsize=(10, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of All Columns')
plt.show()
```

```python
categoric_cols = dataset_31.drop(columns=numeric_cols)
categoric_cols['class'] = dataset_31['class']
```

```python
from scipy.stats import chi2_contingency

chi_squared_values = {}

for col in categoric_cols.columns:
    if col != 'class':  
        contingency_table = pd.crosstab(categoric_cols[col], categoric_cols['class'])
        chi2, _, _, _ = chi2_contingency(contingency_table)
        chi_squared_values[col] = chi2

chi_squared_df = pd.DataFrame(list(chi_squared_values.items()), columns=['Column', 'Chi-squared'])
chi_squared_df = chi_squared_df.sort_values(by='Chi-squared', ascending=False)

plt.figure(figsize=(8, 6))
sns.heatmap(chi_squared_df.set_index('Column'), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('X² correlation between categorical columns and target')
plt.xlabel('Chi-squared value')
plt.ylabel('Categorical column')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
